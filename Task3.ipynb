```python
# Importing necessary libraries
import pandas as pd

# Load datasets
client_df = pd.read_csv(r'E:\PROJECT-AI-ML\BCG X Data Science\client_data (1).csv')
price_df = pd.read_csv(r'E:\PROJECT-AI-ML\BCG X Data Science\price_data (1).csv')

# Displaying first few rows of each dataset to understand their structure
print("Client DataFrame:")
display(client_df.head())

print("Price DataFrame:")
display(price_df.head())

# Removing irrelevant columns with only one unique value

# For client_df
client_single_value_columns = [col for col in client_df.columns if client_df[col].nunique() == 1]
print(f"Columns with a single unique value in client_df: {client_single_value_columns}")
client_df = client_df.drop(columns=client_single_value_columns)

# For price_df
price_single_value_columns = [col for col in price_df.columns if price_df[col].nunique() == 1]
print(f"Columns with a single unique value in price_df: {price_single_value_columns}")
price_df = price_df.drop(columns=price_single_value_columns)

# Display updated DataFrames
print("Updated Client DataFrame:")
display(client_df.head())

print("Updated Price DataFrame:")
display(price_df.head())

# Expanding datasets with new features

# For client_df
if 'date_activ' in client_df.columns:
    client_df['date_activ'] = pd.to_datetime(client_df['date_activ'])
    client_df['join_month'] = client_df['date_activ'].dt.month
    client_df['join_day'] = client_df['date_activ'].dt.day
    client_df['join_year'] = client_df['date_activ'].dt.year

print("Client DataFrame with expanded features:")
display(client_df.head())

# For price_df
if 'price_date' in price_df.columns:
    price_df['price_date'] = pd.to_datetime(price_df['price_date'])
    price_df['price_month'] = price_df['price_date'].dt.month
    price_df['price_day'] = price_df['price_date'].dt.day
    price_df['price_year'] = price_df['price_date'].dt.year

print("Price DataFrame with expanded features:")
display(price_df.head())

# Combining columns

# For client_df
if 'cons_12m' in client_df.columns and 'forecast_cons_12m' in client_df.columns:
    client_df['cons_forecast_ratio'] = client_df['forecast_cons_12m'] / client_df['cons_12m']

print("Client DataFrame with combined features:")
display(client_df.head())

# For price_df
if 'price_off_peak_fix' in price_df.columns and 'price_peak_fix' in price_df.columns:
    price_df['avg_price'] = (price_df['price_off_peak_fix'] + price_df['price_peak_fix']) / 2

print("Price DataFrame with combined features:")
display(price_df.head())

# Merging datasets
merged_df = pd.merge(client_df, price_df, on='id', how='inner')

print("Merged DataFrame:")
display(merged_df.head())
```
