```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Shows plots in jupyter notebook
%matplotlib inline

# Set plot style
sns.set(color_codes=True)
```

```python
import os
files = os.listdir(r'E:\PROJECT-AI-ML\BCG X Data Science')
print(files)
```

```python
client_df = pd.read_csv(r'E:\PROJECT-AI-ML\BCG X Data Science\client_data (1).csv')
price_df = pd.read_csv(r'E:\PROJECT-AI-ML\BCG X Data Science\price_data (1).csv')
```

```python
# Display the first few rows of each dataset
print("Client Data:")
print(client_df.head())

print("\nPrice Data:")
print(price_df.head())
```

```python
# Check data types of columns
print("Client Data Types:")
print(client_df.dtypes)

print("\nPrice Data Types:")
print(price_df.dtypes)
```

```python
# Descriptive statistics for client data
print("Client Data Descriptive Statistics:")
print(client_df.describe(include='all'))

# Descriptive statistics for price data
print("\nPrice Data Descriptive Statistics:")
print(price_df.describe(include='all'))
```

```python
# Check for missing values in client data
print("Missing Values in Client Data:")
print(client_df.isnull().sum())

# Check for missing values in price data
print("\nMissing Values in Price Data:")
print(price_df.isnull().sum())
```

```python
# Plot distribution of numeric columns in client data
client_df.hist(bins=30, figsize=(15, 10))
plt.suptitle("Distribution of Numeric Columns in Client Data")
plt.show()

# Plot distribution of numeric columns in price data
price_df.hist(bins=30, figsize=(15, 10))
plt.suptitle("Distribution of Numeric Columns in Price Data")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot for client data with relevant columns
sns.pairplot(client_df[['cons_12m', 'cons_last_month', 'forecast_price_energy_peak', 'churn']])
plt.suptitle("Pairplot for Selected Columns in Client Data")
plt.show()

# Define the numeric columns
numeric_columns = ['price_off_peak_var', 'price_peak_var', 'price_mid_peak_var', 
                    'price_off_peak_fix', 'price_peak_fix', 'price_mid_peak_fix']

# Select only numeric columns
numeric_price_df = price_df[numeric_columns]

# Compute and plot the correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_price_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap for Price Data")
plt.show()
```
